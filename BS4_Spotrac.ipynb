```python
#Initial scraping of data from spotrac

import requests
import csv

from bs4 import BeautifulSoup
from csv import writer

url = 'https://www.spotrac.com/nfl/rankings/base/'
data = {
    'ajax': 'true',
    'mobile': 'false'
}

soup = BeautifulSoup(requests.post(url, data=data).content, 'html.parser')


#pass contents of website using bs and convert to txt
#soup = BeautifulSoup(response,"html.parser")

#use find all function to scrape positions in the spotrac table
players = soup.find_all("a", class_ = "team-name")
player_list = []
for player in players:
    player_list.append(player.text)
    

#use find all function to scrape positions in the spotrac table
positions = soup.find_all("td", class_ = "center med") 
position_list = []
for position in positions:
    position_list.append(position.text)
    

for i in range(len(position_list)):
    position = position_list[i].replace("\n", "")
    position = position.strip()
    position_list[i] = position
    


    
#use find all function to scrape salaries in the spotrac table
salaries = soup.find_all("span", class_ = "info")
#create empty sal list 
#add salary to empty list
#loop and add elements to sal_list
#sal = []
#for salary in salaries:
#    sal.append(salary.text)
#    i = 0
#    sal_list = []
#    for s in sal:
#        try:
#            sal_list.append(sal[i])
#            i +=7
#        except:
#            break

salary_list = []
for s in salaries:
    salary_list.append(s.text)

for i in range(len(salary_list)):
    salary = salary_list[i].replace("$", "")
    salary = salary.strip()
    salary = salary.replace(",", "")
    salary_list[i] = int(salary)
#create master list of all player data    
list_of_all = [player_list,position_list,salary_list]

#test list by printing contents


#open csv file
#justfor commit
#write to csv file
#create headers
#loop using writefow function   
with open ("salary_Proj.csv", "w") as csvFile:
    csv_writer = writer(csvFile)
    headers = ["Player","Position", "Base Salary"]
    csv_writer.writerow (headers)
    for  i in range(1000):
        csv_writer.writerow ([player_list[i],position_list[i],salary_list[i]])
```
